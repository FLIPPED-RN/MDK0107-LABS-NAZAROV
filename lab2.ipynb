11.1 Базовый уровень | Вариант 4

Поле 1 - Калорийность 100г продукта |
Поле 2 - Вес продукта в граммах |
Функция обработки полей - Вычислить общую калорийность продукта

```python
from datetime import datetime


class Product:
    def __init__(self, calorie_per_100g: float, weight: float):
        self.calorie_per_100g = calorie_per_100g
        self.weight = weight

    def total_calorie(self) -> float:
        return self.calorie_per_100g * self.weight / 100


class Vegetable:
    def __init__(self, name: str, gram: float, calorie_100g: float):
        self.name = name
        self.gram = gram
        self.calorie_100g = calorie_100g

cucumber = Vegetable("огурец", 100, 14)
tomato = Vegetable("помидор", 100, 23)

total_weight = cucumber.gram + tomato.gram

avg_calorie_100g = (
    cucumber.calorie_100g * cucumber.gram +
    tomato.calorie_100g * tomato.gram
) / total_weight

salad = Product(avg_calorie_100g, total_weight)

print("общая калорийность салата:", salad.total_calorie())
```

11.1 Средний уровень | Вариант 4

Класс родитель и его поля - Дата (три числа): часы, минуты, секунды | Функция-метод 1 обработки данных - Вычислить количество секунд в указанном времени | Функция-метод 2 обработки данных - Увеличиить время на 5 секунд

```python
class Date:
    def __init__(self,hours:int,minutes:int,seconds:int):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds

    def to_seconds(self) -> int:
        return self.hours * 3600 + self.minutes * 60 + self.seconds

    def add_five_seconds(self):
        total_seconds = self.to_seconds() + 5

        self.hours = total_seconds // 3600
        total_seconds %= 3600
        self.minutes = total_seconds // 60
        self.seconds = total_seconds % 60

time = Date(1, 59, 58)

print(time.to_seconds())

time.add_five_seconds()
print(time.hours, time.minutes, time.seconds)
```

11.2 Базовый уровень | Вариант 4

Поле класса-потомка - Количество витамина С в 1 грамме продукта | Функция обработки данных - Количество витамина С в продукте

```python
class Product:
    def __init__(self, name:str, weight:float):
        self.name=name
        self.weight=weight

class VitaminCProduct(Product):
    def __init__(self,name:str,weight:float,vitamin_c_per_gram:float):
        super().__init__(name, weight)
        self.vitamin_c_per_gram=vitamin_c_per_gram

    def total_vitamin_c(self)->float:
        return self.weight*self.vitamin_c_per_gram

orange=VitaminCProduct("апельсин", 150, 0.00053)

print("количество витамина С в продукте: ", orange.total_vitamin_c(), "r")
```

11.2 Средний уровень | Вариант 4

Класс-родитель и его поля - Время (три числа): часы, минуты, секунды | Класс-потомок и его поля (поля класса-родителя выделены курсивом) - Расписание движения поездов: номер поезда, направление, время отправления | Функция-метод обработки данных объекта класс-потомка - Количество минут до отправления поезда с указанным номером и введенное время

```python
class Time:
    def __init__(self,hours:int,minutes:int,seconds:int):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds

    def to_seconds(self)->int:
        return self.hours * 3600 + self.minutes * 60 + self.seconds

    def __str__(self):
        return f"{self.hours:02d}:{self.minutes:02d}:{self.seconds:02d}"

class Train(Time):
    def __init__(self, number: str, direction: str, hours:int, minutes:int, seconds: int):
        super().__init__(hours,minutes,seconds)
        self.number = number
        self.direction = direction

    def minutes_before_departure(self, current_time: Time)->int:
        train_seconds=self.to_seconds()
        current_seconds=current_time.to_seconds()

        diff_seconds=train_seconds-current_seconds

        if diff_seconds < 0:
            return 0

        return diff_seconds // 60

current_time = Time(10,30,0)

train=Train("125A", "Москва - Санкт-Петербург", 11,0,0)
print("номер поезда: ", train.number)
print("отправление поезда через: ", train.minutes_before_departure(current_time), "минут")
print("время сейчас ", current_time)
```

11.3 Базовый уровень | Вариант 4

Кабель: тип, количество жил кабеля, диаметр. Q = диаметр / количество жил
P: год выпуска товара Qp = Q + 0,5 * (T-P), где T - текущий год

```python
class Cable:
    def __init__(self, type:str,numberCableCores:int,diameter:float):
        self.type = type
        self.numberCableCores = numberCableCores
        self.diameter = diameter

    def Q(self)->float:
        return self.diameter/self.numberCableCores

class Product(Cable):
    def __init__(self, type:str,numberCableCores:int,diameter:float,year:int):
        super().__init__(type,numberCableCores,diameter)
        self.year=year

    def Qp(self)->float:
        T=datetime.now().year
        return self.Q()+0.5*(T-self.year)

cable=Product("силовой",4,8.0,2020)

print("ип кабеля:",cable.type)
print("Q=",cable.Q())
print("qp=",cable.Qp())
```
